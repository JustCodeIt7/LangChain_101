```python
from langchain_ollama import OllamaEmbeddings
from dotenv import load_dotenv
load_dotenv()

text = """Cats are known for their quirky personalities and playful antics. They can go from being aloof to affectionate in a matter of seconds."""

len(text)
```

# Ollama Embeddings

```python
embedding_model = OllamaEmbeddings(
    model="mxbai-embed-large",
)


# Get the embeddings
embeddings = embedding_model.embed_documents([text,text])

len(embeddings)
```

```python
embeddings[0]
len(embeddings[0])
```

```python
# embedding query
query = "Cats are known for their quirky personalities and playful antics."

query_embedding = embedding_model.embed_query(query)

print(query_embedding)
```

# OpenAI Embeddings

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings()

# Get the embeddings
embeddings = embedding_model.embed_documents([text])

len(embeddings)
```
